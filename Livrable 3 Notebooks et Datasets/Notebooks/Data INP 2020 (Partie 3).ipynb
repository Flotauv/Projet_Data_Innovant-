Construction d'un nouveau dataset pour analyser les mobilités douces :

L'objectif est de créer un nouveau dataset permettant deux types d'analyses :
- Visualisation en secteurs (diagramme circulaire) : Représenter la répartition de l'usage des mobilités douces par les usagers.
- Corrélation : Établir une relation entre la distance moyenne parcourue par un usager pour un mode de transport donné et le choix de ce mode (mesuré par les occurrences des usagers pour chaque mode de transport).

```python
comptage_mode_transport_A = df.iloc[:, 10:21].count()
comptage_mode_transport_2020 = comptage_mode_transport_A.reset_index()
comptage_mode_transport_2020.columns = ['Mode de Transport 2020', 'Occurrences']
comptage_mode_transport_2020['Mode de Transport 2020'] = comptage_mode_transport_2020['Mode de Transport 2020'].str.replace(r'^\d+\.\s*', '', regex=True)
somme_mode_transport = df.iloc[:, 10:21].fillna(0).sum()
comptage_mode_transport_2020['Distance Totale (km)'] = somme_mode_transport.values
comptage_mode_transport_2020['Distance Moyenne (km)'] = (
    comptage_mode_transport_2020['Distance Totale (km)'] /
    comptage_mode_transport_2020['Occurrences']
)
comptage_mode_transport_2020
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 8))  # Taille du graphique
plt.pie(
    comptage_mode_transport_2020['Occurrences'],  # Les données (valeurs)
    labels=comptage_mode_transport_2020['Mode de Transport 2020'],  # Les étiquettes (modes de transport)
    autopct='%1.1f%%',  # Affiche les pourcentages avec 1 chiffre après la virgule
    startangle=90,  # Angle de départ pour un affichage propre
    wedgeprops={'edgecolor': 'black'}  # Bordures noires pour les secteurs
)
plt.title("Répartition des modes de transport en 2020")
plt.show()
```

Pour les répondants de Grenoble INP, les mobilités douces sont représentées à 50%. Il faut aussi prendre en compte
l'événement Covid de l'année qui peut avoir une lourde influence sur cette répartition.
