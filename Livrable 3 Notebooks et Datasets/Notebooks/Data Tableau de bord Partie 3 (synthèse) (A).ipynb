```python
import pandas as pd
df1 = pd.read_excel('Data INP/comptage_mode_transport_2020.xlsx')
df2 = pd.read_excel('Data INP/comptage_mode_transport_2021.xlsx')
df3 = pd.read_excel('Data INP/comptage_mode_transport_2023.xlsx')
df4 = pd.read_excel('Data INP/comptage_mode_transport_2024.xlsx')
```

```python
df1.columns = ['Mode de Transport', 'Occurrences 2020', 'Distance Totale (km) 2020', 'Distance Moyenne (km) 2020']
total_occurrences = df1['Occurrences 2020'].sum()
# Les proportions (en pourcentage)
df1['Proportion des occurences 2020'] = (df1['Occurrences 2020'] / total_occurrences) * 100
df1['Distance Moyenne Pondérée 2020'] = df1['Distance Moyenne (km) 2020'] * (df1['Proportion des occurences 2020'] / 100)
df1
```

```python
df2.columns = ['Mode de Transport', 'Occurrences 2021', 'Distance Totale (km) 2021', 'Distance Moyenne (km) 2021']
total_occurrencesB = df2['Occurrences 2021'].sum()
# Les proportions (en pourcentage)
df2['Proportion des occurences 2021'] = (df2['Occurrences 2021'] / total_occurrencesB) * 100
df2['Distance Moyenne Pondérée 2021'] = df2['Distance Moyenne (km) 2021'] * (df2['Proportion des occurences 2021'] / 100)
df2
```

```python
df3.columns = ['Mode de Transport', 'Occurrences 2023', 'Distance Totale (km) 2023', 'Distance Moyenne (km) 2023']
total_occurrencesC = df3['Occurrences 2023'].sum()
# Les proportions (en pourcentage)
df3['Proportion des occurences 2023'] = (df3['Occurrences 2023'] / total_occurrencesC) * 100
df3['Distance Moyenne Pondérée 2023'] = df3['Distance Moyenne (km) 2023'] * (df3['Proportion des occurences 2023'] / 100)
df3
```

```python
df4.columns = ['Mode de Transport', 'Occurrences 2024', 'Distance Totale (km) 2024', 'Distance Moyenne (km) 2024']
total_occurrencesD = df4['Occurrences 2024'].sum()
# Les proportions (en pourcentage)
df4['Proportion des occurences 2024'] = (df4['Occurrences 2024'] / total_occurrencesD) * 100
df4['Distance Moyenne Pondérée 2024'] = df4['Distance Moyenne (km) 2024'] * (df4['Proportion des occurences 2024'] / 100)
df4
```

```python
df_fusion = df1.merge(df2, on='Mode de Transport') \
               .merge(df3, on='Mode de Transport') \
               .merge(df4, on='Mode de Transport')
df_fusion
df_fusion.to_excel('Data INP/datafusion.xlsx', index=False)
```

```python
import matplotlib.pyplot as plt



# Transformation du DataFrame en format long pour faciliter la visualisation
df_melted = df_fusion.melt(
    id_vars='Mode de Transport',  # Colonne contenant les modes de transport
    value_vars=[
        'Proportion des occurences 2020', 
        'Proportion des occurences 2021', 
        'Proportion des occurences 2023', 
        'Proportion des occurences 2024'
    ],
    var_name='Année',
    value_name='Proportion'
)

# Nettoyage des noms des années pour extraire uniquement l'année
df_melted['Année'] = df_melted['Année'].str.extract(r'(\d{4})')

# Visualisation
plt.figure(figsize=(12, 8))

#  Une ligne pour chaque mode de transport
for mode in df_melted['Mode de Transport'].unique():
    subset = df_melted[df_melted['Mode de Transport'] == mode]
    plt.plot(subset['Année'], subset['Proportion'], marker='o', label=mode)

# Titre, les labels, et la légende
plt.title("Évolution des proportions des occurrences par mode de transport (2020-2024)", fontsize=14)
plt.xlabel("Année", fontsize=12)
plt.ylabel("Proportion des Occurrences (%)", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title="Mode de Transport", fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Nous pouvons discerner trois tendances. La première: à la hausse, nous avons (tram, pied, trotinette ou vélo). Ce sont les modes de transport
les plus utilisés. On peut faire le lien avec le très bon réseau d'infrastructures de Grenoble. Puis, nous avons les transports stagnants: Bus, train, vélo électrique et covoiturage. Le train est utilisé surtout pas les personnes habitants dans la périphérie de Grenoble, il s'agit du seul mode de transport disponible pour les très longues distances avec la voiture. Ce transport est amené à stagner. De plus, il n'y a que trois gares dans la ville de Grenoble. Concernant le bus, ce moyen de transport peut être vu comme plus chronophage que le vélo par exemple. Enfin, le covoiturage est très peu développé en France et le vélo électrique est encore onereux. Enfin, conformément aux aménagements urbains de la ville de Grenoble, il est de plus en plus difficile de circuler en ville avec les modes de transport traditionnels comme la voiture. D'où le déclin.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Transformation des données pour la visualisation
df_melted = df_fusion.melt(
    id_vars='Mode de Transport',  # Column for modes of transport
    value_vars=[
        'Distance Moyenne Pondérée 2020', 
        'Distance Moyenne Pondérée 2021', 
        'Distance Moyenne Pondérée 2023', 
        'Distance Moyenne Pondérée 2024'
    ],
    var_name='Année',
    value_name='Distance Moyenne Pondérée'
)

# Nettoyage des noms des années pour extraire uniquement l'année
df_melted['Année'] = df_melted['Année'].str.extract(r'(\d{4})')

# Visualisation
plt.figure(figsize=(12, 8))

#  Une ligne pour chaque mode de transport
for mode in df_melted['Mode de Transport'].unique():
    subset = df_melted[df_melted['Mode de Transport'] == mode]
    plt.plot(subset['Année'], subset['Distance Moyenne Pondérée'], marker='o', label=mode)

# Titre, les labels, et la légende
plt.title("Évolution des distances moyennes pondérées par mode de transport (2020-2024)", fontsize=14)
plt.xlabel("Année", fontsize=12)
plt.ylabel("Distance Moyenne Pondérée (km)", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.legend(title="Mode de Transport", fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

Avertissement: l'année 2024 présente des années abérrantes (bus et pied). Nous ne commenterons pas cette année sur bus et pied. Nous avons la distance moyenne parcourue par les usagers sur chaque mode de transport. Les moyennes sont pondérées par la proportion d'usagers sur
le moyen de transport. Si on se réfère aux distances, le train, le vélo, le bus, et le tram sont les modes de transport englobant le plus de distance. 
La voiture est utilisée pour très peu de distance. Tandis que le coviturage et le vélo électrique ont des positions intermédiaires. Si nous généralisons ces observations à l'ensemble de la ville, nous pouvons dire que les trajets grenoblois sont largement décarbonés. Les politiques de la ville semble largement faciliter la vie des modes de transport doux. Il peut y avoir des biais sur le public cible des enquêtes de l'INP. En effet, il s'agit d'un public "académique", un groupe qui peut sembler être plus acquis aux causes environnementales ou plus démunis privilégiant donc les transports moins onéreux, donc doux.
