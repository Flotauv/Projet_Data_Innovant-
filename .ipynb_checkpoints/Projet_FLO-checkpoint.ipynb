## Base de données accidents 

```python
df_accidents = pd.read_csv("BaseDeDonnées/Accidents_france_2005_2016/caracteristics.csv",encoding='latin-1')
df_accidents
```

```python
df_vehicules = pd.read_csv('BaseDeDonnées/Accidents_france_2005_2016/vehicles.csv')
df_vehicules
```

```python
df_users = pd.read_csv('BaseDeDonnées/Accidents_france_2005_2016/users.csv')

df_users
```

```python
df_accidents.dtypes
```

```python
df_accidents_grenoble = df_accidents[(df_accidents['dep']== 380) & (df_accidents['gps']=='M')]

df_accidents_grenoble
```

```python
df_accidents_grenoble = pd.merge(df_accidents_grenoble,df_vehicules, how='left',on='Num_Acc')
```

```python
df_accidents_grenoble = df_accidents_grenoble[df_accidents_grenoble['catv']==1]

df_accidents_grenoble
```

```python
df_accidents_grenoble = pd.merge(df_accidents_grenoble,df_users, how='left',on='Num_Acc')
```
